### Problema 1

```python
### a) e b)
L = 1.0      
C = 1.0     
d = 0.1     
N = 100      
sigma = 0.3  
a = L/N      
v = 100.0   
h = 1e-6     
epsilon = h/100
```

```python
x = np.linspace(0, L, N)

phi = np.zeros(N, float)
psi = np.zeros(N, float)
```

```python
def initial(x):
    return (C*x*(L-x)/(L**2))*np.exp((-(x-d)**2)/(2*sigma**2))
```

```python
for i in range(N):
    psi[i] = initial(x[i])
    
new_phi = np.copy(phi)    
new_psi = np.copy(psi)
```

```python
t_val = [25e-3,50e-3,75e-3,100e-3]
t = 0.0
```

```python
while (t<(t_val[-1]+epsilon)):
    ## euler method
    for i in range(1, N-1):
        new_phi[i] = phi[i] + h*psi[i]
        new_psi[i] = psi[i] + h*(phi[i-1]+phi[i+1]-2*phi[i])*(v**2)/(a**2)
    t+=h
    
    phi = np.copy(new_phi)
    psi = np.copy(new_psi)
    
    if (abs(t-t_val[0])<=epsilon) or (abs(t-t_val[1])<=epsilon) or (abs(t-t_val[2])<=epsilon) or (abs(t-t_val[3])<=epsilon):
        plt.plot(x,phi,label="t = %.4f" % t)
        plt.legend()
        plt.show()
```

### Problema 2

```python
g = 9.81
N = 100 
ti = 0 
tf = 10 
h = (tf-ti)/N 
tempos = np.linspace(ti,tf,N) 
x = np.zeros(N, float) 

erro = 1 
tol = 10e-6
```

```python
while erro>tol:
    aux = x.copy()
    for i in range(1,N-1):
        x[i] = 0.5*(g*h*h + aux[i-1] + aux[i+1]) 
    erro = np.max(abs(x-aux)) 
    

print(f'A altura mÃ¡xima alcanÃ§ada Ã© de {max(x):.2f} m')    
    
plt.plot(tempos,x)
plt.xlabel("t [s]")
plt.ylabel("x [m]")
plt.show()
```
